```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy
from scipy import signal
import numpy as np
sns.set()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/stockdata.csv')
df.head()
```

```python
sample = range(15)
saw = signal.sawtooth(t=sample)

data_sample = list(df['SBUX'][0:100])
data_sample2 = list(df['AAPL'][0:100])
x = list(range(len(data_sample)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(x,data_sample,label='SBUX')
plt.plot(x,data_sample2,label='AAPL')
y_convolve = signal.convolve(saw, data_sample2)
x_convolve = list(range(len(y_convolve)))
plt.plot(x_convolve,y_convolve,label='conv AAPL')
y_convolve = signal.convolve(saw, data_sample)
x_convolve = list(range(len(y_convolve)))
plt.plot(x_convolve,y_convolve,label='conv SBUX')
plt.legend()
plt.show()
```
